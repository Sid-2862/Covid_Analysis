# 2. Data Cleaning:

```python
Handle missing values.
Convert data types if necessary.
Check for duplicates.
```

```python
# Handle missing values



```

```python
# check missing values  isna()  / isnull()
df.isna().sum()
```

```python
#treating the missing values
cat=[]
con=[]
for i in df:
    if df[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
print(cat)
print(con)
```

```python
# for categorical column repace null values with mode
for i in cat:
    print(f"Modal value of {i} is {df[i].mode()[0]}")
```

```python
#ffill up missing values fillna()

for i in df:
    if df[i].dtypes=='object':
        m=df[i].mode()[0]
        df[i]=df[i].fillna(m)
```

```python
df.isna().sum()
```

# 4. Time Series Analysis:

Analyze the trends in confirmed, deaths, and recovered cases over time.<br>
Create time-based visualizations.

```python
# Time series analysis
# groupby 'ObservationDate' and apply aggregation functions like sum on deaths,confirmed ,recovered 
# df_time = df.groupby('ObservationDate').agg({'Confirmed': 'sum', 'Deaths': 'sum', 'Recovered': 'sum'})

df1=df.groupby('ObservationDate').agg({'Deaths':'sum','Confirmed':'sum','Recovered':'sum'})
df1

```

```python
# Plot time series data using line plot
import matplotlib.pyplot as plt
import seaborn as sns
# syntax: 
#sns.lineplot(data=df_time, x='ObservationDate', y='Confirmed', label='Confirmed')
# sns.lineplot(data=df_time, x='ObservationDate', y='Deaths', label='Deaths')
```

```python
plt.figure(figsize=(5,5))
sns.lineplot(data=df1,x='ObservationDate',y='Deaths',label='Deaths')
sns.lineplot(data=df1,x='ObservationDate',y='Confirmed',label='Confirmed')
sns.lineplot(data=df1,x='ObservationDate',y='Recovered',label='Recovered')
plt.title('Line PLot showing Deaths,Confirmed,Recovered')
plt.show()
```

# 5. Geographical Analysis:

Analyze the distribution of cases across different countries/regions.<br>
Create geographical visualizations (e.g., world map).

```python
# Geographical analysis
# df_country = df.groupby('Country/Region').agg({'Confirmed': 'sum', 'Deaths': 'sum', 'Recovered': 'sum'})

df2=df.groupby('Country/Region').agg({'Deaths':'sum','Confirmed':'sum','Recovered':'sum'})

df2
```

```python
# Plot geographical data
# sns.barplot(x=df_country.index, y='Confirmed', data=df_country)
```

```python
plt.figure(figsize=(50,60))
sns.barplot(x='Confirmed' ,y=df2.index ,data=df2)
sns.barplot(x='Deaths' ,y=df2.index ,data=df2)
sns.barplot(x='Recovered' ,y=df2.index ,data=df2)

plt.show()
```

# 6. Correlation Analysis:

Examine the correlation between different numeric variables.

```python
# Correlation analysis between confirmed,deaths and recovered
cdr=df[['Confirmed','Deaths','Recovered']].corr()
cdr

```

```python
# Plot correlation on heat map
plt.figure(figsize=(5,5))
sns.heatmap(cdr,annot=True,cmap='Reds')
plt.title('Heatmap showing Correlation betweem Deaths,Recovered,Confirmed')
plt.show()
```
